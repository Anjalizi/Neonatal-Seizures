```python
X = data.drop(columns=['Signal','Label','Hurst Component'])
y = data['Label']
```

```python
cols = X.columns
data = pre.scale(X)
data = pd.DataFrame(data=data[0:,0:], index=data[0:,0], columns=data[0,0:]) 
data.columns = cols
arr = []
for num in range(1400):
    arr.append(num);
data = data.set_index(pd.Index(arr))
data
```

```python
from sklearn.feature_selection import VarianceThreshold
def variance_threshold_selector(data, threshold):
    selector = VarianceThreshold(threshold)
    selector.fit_transform(data)
    return data[data.columns[selector.get_support(indices=True)]]
```

```python
p = 0.8
X = variance_threshold_selector(X, p*(1-p))
X.shape
```

```python
cols = X.columns
```

```python
data = pre.scale(X)
data = pd.DataFrame(data=data[0:,0:], index=data[0:,0], columns=data[0,0:]) 
data.columns = cols
arr = []
for num in range(1400):
    arr.append(num);
data = data.set_index(pd.Index(arr))
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
x_val, x_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=0)
```

```python
from sklearn import preprocessing as pre
X_train = pre.scale(X_train)
x_test = pre.scale(x_test)
x_val = pre.scale(x_val)
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
parameters = {'n_neighbors':range(2,25), 'weights':('uniform', 'distance'), 'algorithm':('auto', 'ball_tree', 'kd_tree', 'brute')}
knn = KNeighborsClassifier()
clf = GridSearchCV(knn, parameters)
```

```python
clf.fit(X_train, y_train)
```
